### Airbnb Market Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("airbnb.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all").T
```

## Data Cleaning

```python
# Drop unnecessary columns
df.drop(columns=['Unnamed: 0', 'id', 'host_id', 'img_links'], inplace=True)
```

```python
# Fix "rating" Column
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'].isna().sum()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Handle missing values in "rating" column
df['rating'] = df['rating'].fillna(df['rating'].median())
df['rating'].isna().sum() 
```

```python
# Handle missing values in "checkin" column
df['checkin'] = df['checkin'].fillna(df['checkin'].mode()[0])
df['checkin'].isna().sum() 
```

```python
# Handle missing values in "checkout" column
df['checkout'] = df['checkout'].fillna(df['checkout'].mode()[0])
df['checkout'].isna().sum() 
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Change "price" column to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df[df['price'].notna()]
```

```python
# new column "price_per_guest"
df['price_per_guest'] = df['price'] / df['guests']
```

```python
df.head()
```

## EDA

```python
# Top 10 Countries by Listing Count
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Top 10 Countries by Airbnb Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Country')
plt.show()
```

```python
# Price Distribution
plt.figure(figsize=(12,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show()
```

```python
#  Average price by country
avg_price = df.groupby('country')['price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=avg_price.values, y=avg_price.index, palette='magma')
plt.title('Average Airbnb Price by Country')
plt.xlabel('Average Price')
plt.ylabel('Country')
plt.show()
```

```python
# Rating vs Price Scatterplot
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='rating', y='price', alpha=0.6)
plt.title('Rating vs Price')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

```python
# Avalability based on Checkout Time
top_checkout = df['checkout'].value_counts().head(7)
plt.figure(figsize=(12,6))
sns.barplot(x=top_checkout.index, y=top_checkout.values, palette='magma')
plt.title('Most Common Checkout Times')
plt.xlabel('Checkout Time')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

```python
# New Cleaned Dataset
df.to_csv("airbnb_cleaned.csv", index=False)
```
